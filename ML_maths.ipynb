```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear_fxn_1D(x,b,w):
    y = np.dot(w,x) + b
    return y
```

```python
x = np.arange(0.0,10.0,0.01)
b = 5.0
w = 4.0
y = linear_fxn_1D(x,b,w)
```

```python
# plotting the function
fig, ax = plt.subplots()
ax.plot(x,y,'r-')
ax.set_ylim([0,40])
ax.set_xlim([0,10])
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

```python
def linear_fxn_2D(x1,x2,w1,w2,b):
    y = np.dot(w1,x1) + np.dot(w2,x2) + b
    return y
```

```python
def draw_2D_fxn(x1_mesh, x2_mesh, y):
    fig,ax = plt.subplots()
    fig.set_size_inches(7,7)
    pos = ax.contourf(x1_mesh, x2_mesh, y, levels = 256, cmap='hot', vmin=10, vmax=10.0)
    fig.colorbar(pos, ax = ax)
    ax.set_xlabel('x1')
    ax.set_ylabel('y')
    levels = np.arange(-10, 10, 1.0)
    ax.contour(x1_mesh, x2_mesh, y, levels, cmap = 'winter')
    plt.show()
```

```python
x1 = np.arange(0.0, 10.0, 0.1)
x2 = np.arange(0.0, 10.0, 0.1)
x1,x2 = np.meshgrid(x1,x2)

b = 5.0
w1 = 1.0
w2 = -0.5
y = linear_fxn_2D(x1,x2,w1,w2,b)
draw_2D_fxn(x1,x2,y)
```
